## Build a recommender system by using cosine simillarties score.

### Import necessary lib

```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
```

### Import dataset

```python
book_data = pd.read_csv('book (1).csv')
book_data
```

```python
book_data1 = book_data.rename(columns={'User.ID':'UserID','Book.Title':'Book_title','Book.Rating':'Book_rating'})
book_data1.head()
```

```python
book_data1.Book_title.nunique()
```

```python
book_data1.sort_values(['UserID'])
```

```python
# convert a data using pivot table
book_recom = pd.pivot_table(data = book_data1,values='Book_rating',index='UserID',columns='Book_title').reset_index(drop=True)
book_recom
```

```python
book_recom.index = book_data1['UserID'].unique()
book_recom
```

```python
book_recom1 = book_recom.fillna(0)
book_recom1
```

### calculate cosine similarity between users

```python
user_sim =1-pairwise_distances(X = book_recom1.values,metric='cosine')
user_sim
```

```python
user_sim_df = pd.DataFrame(data=user_sim)
user_sim_df
```

```python
user_sim_df.index = book_data1['UserID'].unique()
user_sim_df.columns = book_data1['UserID'].unique()
user_sim_df
```

```python
# fill diagonal value with 0
np.fill_diagonal(user_sim_df.to_numpy(),0)
user_sim_df
```

### Check for most similarity user

```python
x = user_sim_df.idxmax(axis=1)
x
```
